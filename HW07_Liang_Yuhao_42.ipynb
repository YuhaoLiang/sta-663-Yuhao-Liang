### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probability of a rat dying. We are given $n = 5$, but we believe that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
np_df = np.array([[-0.896,5,0],[-0.296,5,1],[-0.053,5,3],[0.727,5,5]])
df = pd.DataFrame(np_df, columns=['Dose', 'Rats', 'Deaths'])
df[['Rats']] = df[['Rats']].astype(int)
df[['Deaths']] = df[['Deaths']].astype(int)
import pystan

log_reg_code = """
data {
    int<lower=0> m; // data size
    int rats[m]; // number of rats in one experiement 
    int deaths[m]; // number of deaths
    real dose[m]; // dose quantity
}
transformed data {}
parameters {
    real alpha;
    real beta;
}
transformed parameters {}
model {
    alpha ~ normal(0, 5);
    beta ~ normal(0, 10);
    for(i in 1:m) 
    deaths[i] ~ binomial_logit(rats[i],alpha + beta*dose[i]);
}
generated quantities {}
"""

log_reg_dat = {
             'm': len(df),
             'rats': df.Rats,
             'deaths': df.Deaths,
             'dose': df.Dose,
            }

fit = pystan.stan(model_code=log_reg_code, data=log_reg_dat, iter=2000, chains=1)
```

```python
print(fit)
```

### Using a hierarchcical model

This uses the Gelman radon data set and is based off this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the preence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
radon.head()
```

```python
county_dict = set(radon.county)
for i, s in enumerate(county_dict):
    radon.ix[radon.county == s,'county'] = i + 1
radon[['county']] = radon[['county']].astype(int)
radon[['floor']] = radon[['floor']].astype(int)
```

```python
lin_reg_code = """
data {
    int<lower=0> n; //data size 
    int<lower=0> m; //number of county
    int county[n];
    int floor_ind[n];
    real y[n]; 
}
transformed data {}
parameters {
    real a[m];
    real mu_a;
    real<lower=0> sigma_a;
    real b[m];
    real mu_b;
    real<lower=0> sigma_b;
    real sigma;
}
transformed parameters {}
model {
    for (i in 1:m){
        a[i] ~ normal(mu_a,sigma_a);
        b[i] ~ normal(mu_b,sigma_b);
        }
    mu_a ~ normal(0,10);
    sigma_a ~ uniform(0, 20);
    mu_b ~ normal(0,10);
    sigma_b ~ uniform(0, 20);
    sigma ~ uniform(0, 20);
    for (j in 1:n){
        y[j] ~ normal(a[county[j]] + b[county[j]]*floor_ind[j], sigma);
        }
}
generated quantities {}
"""

lin_reg_dat = {
             'n': len(radon),
             'm': len(county_dict),
             'floor_ind': radon.floor,
             'county': radon.county,
             'y': radon.log_radon,
            }

fit = pystan.stan(model_code=lin_reg_code, data=lin_reg_dat, iter=1000, chains=1)
```

```python
print(fit)
```

**Exercise 3 (10 points)**: Implement the leap-frog method and simulate the spring dynamics following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the spring trajectory plots.

```python

# EXAMPLE 1: SIMULATING HAMILTONIAN DYNAMICS
#            OF HARMONIC OSCILLATOR
# STEP SIZE
delta = 0.1;
 
# LEAP FROG
S = 70;

# DEFINE KINETIC ENERGY FUNCTION
def kin(p):
    return p**2/2.0
 
# DEFINE POTENTIAL ENERGY FUNCTION FOR SPRING (K =1)
def eng(x):
    return x**2/2.0
 
# DEFINE GRADIENT OF POTENTIAL ENERGY
def grad_eng(x):
    return x
 
# INITIAL CONDITIONS
x0 = -4; # POSTIION
p0 = 1;  # MOMENTUM
plt.figure(figsize=(12, 9));

## SIMULATE HAMILTONIAN DYNAMICS WITH LEAPFROG METHOD
# FIRST HALF STEP FOR MOMENTUM
pStep = p0 - delta/2.0*grad_eng(x0);
 
# FIRST FULL STEP FOR POSITION/SAMPLE
xStep = x0 + delta*pStep;
fig1 = plt.figure();

for j in range(S-1):
    # UPDATE MOMENTUM
    pStep = pStep - delta*grad_eng(xStep);
    
    # UPDATE POSITION
    xStep = xStep + delta*pStep;
    
    ax = plt.subplot(224);
    ax.plot(xStep,pStep,'ko');
    ax.set_xlim([-6,6]); 
    ax.set_ylim([-6,6]);
    ax.set_xlabel('x'); 
    ax.set_ylabel('p');
    ax.set_title('Phase Space');
    
    import time
    time.sleep(0.1)

# (LAST HALF STEP FOR MOMENTUM)
pStep = pStep - delta/2*grad_eng(xStep);
```

**Exercise 4 (20 points)**: Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Basyesian Data Analysis" by Jouh Kruschke.) 

```python
# EXAMPLE 2: HYBRID MONTE CARLO SAMPLING -- BIVARIATE NORMAL
import random
random.seed(123)

# STEP SIZE
delta = 0.3;
nSamples = 1000;
S = 20;

import numpy.linalg as la

sgm = np.array([[1,0.8],[0.8,1]])

# DEFINE POTENTIAL ENERGY FUNCTION
def eng(x):
    return x.T.dot(la.inv(sgm)).dot(x)
 
# DEFINE GRADIENT OF POTENTIAL ENERGY
def grad_eng(x):
    return x.T.dot(la.inv(sgm))
 
# DEFINE KINETIC ENERGY FUNCTION
def kin(p):
    return np.sum(p.T.dot(p))/2.0
 
# INITIAL STATE
x = np.zeros((2,nSamples));
x0 = np.array([0,6]);
x[:,0] = x0;

t = 0;
import numpy.random as nprd
```

```python

while (t < nSamples-1):
    t += 1;
    # SAMPLE RANDOM MOMENTUM
    p0 = nprd.normal((2,1));
    ## SIMULATE HAMILTONIAN DYNAMICS
    # FIRST 1/2 STEP OF MOMENTUM
    pStar = p0 - delta/2.0*grad_eng(x[:,t-1]);
    # FIRST FULL STEP FOR POSITION/SAMPLE
    xStar = x[:,t-1] + delta*pStar;
    # FULL STEPS
    for j in range(S-1):
        # MOMENTUM
        pStar = pStar - delta*grad_eng(xStar);
        # POSITION/SAMPLE
        xStar = xStar + delta*pStar;
    
    # LAST HALP STEP
    pStar = pStar - delta/2*grad_eng(xStar);
    # COULD NEGATE MOMENTUM HERE TO LEAVE
    # THE PROPOSAL DISTRIBUTION SYMMETRIC.
    # HOWEVER WE THROW THIS AWAY FOR NEXT
    # SAMPLE, SO IT DOESN'T MATTER
    
    # EVALUATE ENERGIES AT
    # START AND END OF TRAJECTORY
    U0 = eng(x[:,t-1]);
    UStar = eng(xStar);
    
    K0 = kin(p0);
    KStar = kin(pStar);
    
    # ACCEPTANCE/REJECTION CRITERION
    alpha = min(1,np.exp((U0 + K0) - (UStar + KStar)));
    
    u = nprd.uniform(size=1);
    if u < alpha:
        x[:,t] = xStar;
    else:
        x[:,t] = x[:,t-1];
 
```

```python
# DISPLAY
plt.figure(figsize=(12, 9))
plt.scatter(x[0,:],x[1,:]);
plt.plot(x[0,0:200],x[1,0:200],'ro-');
plt.xlim([-6,6]); 
plt.ylim([-6,6]);
plt.legend();
plt.title('Hamiltonian Monte Carlo');

```
