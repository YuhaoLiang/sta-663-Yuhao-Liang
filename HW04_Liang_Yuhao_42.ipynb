**Exercise 2 (10 points)**. Write 3 functions to calculate the term frequency (tf), the inverse document frequency (idf) and the product (tf-idf). Each function should take a single argument `docs`, which is a dictionary of (key=identifier, value=dcoument text) pairs, and return an appropriately sized array. Convert '-' to ' ' (space), remove punctuation, convert text to lowercase and split on whitespace to generate a collection of terms from the dcoument text.

- tf = the number of occurrences of term $i$ in document $j$
- idf = $\log \frac{n}{1 + \text{df}_i}$ where $n$ is the total number of documents and $\text{df}_i$ is the number of documents in which term $i$ occurs.

Print the table of tf-idf values for the following document collection

```
s1 = "The quick brown fox"
s2 = "Brown fox jumps over the jumps jumps jumps"
s3 = "The the the lazy dog elephant."
s4 = "The the the the the dog peacock lion tiger elephant"

docs = {'s1': s1, 's2': s2, 's3': s3, 's4': s4}
```
Note: You can use either a numpy array or pandas dataframe to store the matrix. However, we suggest using a Pandas dataframe since that will allow you to keep track of the row (term) and column (document) names in a single object. Of course, you could also maintain a numpy matrix, a list of terms, and a list of documents separately if you prefer.

````python
#Your code
s1 = "The quick brown fox"
s2 = "Brown fox jumps over the jumps jumps jumps"
s3 = "The the the lazy dog elephant."
s4 = "The the the the the dog peacock lion tiger elephant"

docs = {'s1': s1, 's2': s2, 's3': s3, 's4': s4}
print docs
import string
# Suggest function signatures
def tf(doc):
    """Returns the number of times each term occurs in a dcoument.
    We preprocess the document to strip punctuation and convert to lowercase.
    Terms are found by splitting on whitespace."""
    #Remove punctuation, convert to lower case and split
    s_temp = doc.lower().translate(None, string.punctuation).replace('-', ' ').replace('\n', ' ').translate(None, string.digits).split()
    #Get the unique value
    s_uni = list(set(s_temp))
    #Generate a list for times of each term
    times = []
    for i in range(len(s_uni)):
        times.append(s_temp.count(s_uni[i]))
    #Return as a pandas dataframe
    df = pd.DataFrame(times,s_uni)
    #Set up column name
    df.columns = ['freq']
    return(df )
    

def tfs(docs):
    """Create a term freqeuncy dataframe from a dictionary of documents."""
    #Get term frequency of each document and merge together
    df = pd.concat([tf(docs[s]) for s in docs.keys()],axis=1)
    #Replace nan with 0
    df[np.isnan(df)] = 0
    #Set up column name
    df.columns = docs.keys()
    return(df)
    

def idf(docs):
    """Find inverse document frequecny series from a dictionry of doucmnets."""
    #Get the dataframe of frequency of docs
    df = tfs(docs)
    #Number of documents
    n = len(df.columns)
    #idf
    data = np.log(n/(1.0+np.array([np.sum(df.loc[s,:])  for s in list(df.index)]))).tolist()
    #return series
    return(pd.Series(data, index=list(df.index)))

def tf_idf(docs):
    """Return the product of the term-frequency and inverse document freqeucny."""
    #Get the term-frequency
    df = tfs(docs)
    #Get the idf
    s = idf(docs)
    #Product of the term-frequency and inverse document freqeucny
    df2 = pd.concat([s[list(df.ix[:,r].index)]*df.ix[:,r]  for r in df.columns],axis=1)
    #Reset columns name
    df2.columns = df.columns
    return(df2)

tf_idf(docs)
````
