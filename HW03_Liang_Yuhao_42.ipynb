**Exercise 1 (10 pts)**. Avoding catastrophic cancellation.

The tail of the standard logistic distributon is given by $1 - F(t) = 1 - (1+e^{-t})^{-1}$.

- Define a function `f1` to calculate the tail probability of the logistic distribution using the formula given above
- Use `sympy` to find the exact value of the tail distribution (using the same symbolic formula) to 20 decimal digits
- Calculate the *relative error* of `f1` when $t = 25$ (The relative error is given by `abs(exact - approximate)/exact`)
- Rewrite the expression for the tail of the logistic distribution using simple algebra so that there is no risk of cancellation, and write a function `f2` using this formula. Calculate the *relative error* of `f2` when $t = 25$. 
- How much more accurate is `f2` compared with `f1` in terms of the relative error?

````python
# Your code here
t = 25
###Function 1
def f1(t):
    return(1.0 - 1.0/(1.0 + np.exp(-t)))

print "f1(25)",'%.20f' % f1(t)

###Using sympy to find exact value
import sympy
x = sympy.Symbol('x')
e = 1 - 1/(1 + sympy.exp(-x))
#substitute x with 25
e = e.subs(x,25)
print "Exact value",'%.20f' % e.evalf()
exact = e.evalf()
###Relative error of function 1
print "Relative error of f1",abs(exact - f1(t))/exact

###Function 2
def f2(t):
    return(1.0/(1.0 + np.exp(t)))

print "f2(25)",'%.20f' % f2(t)

###Relative error of function 2
print "Relative error of f2",abs(exact - f2(t))/exact

###Accuracy comparison
print "f2 is",(abs(exact - f1(t))/exact)/(abs(exact - f2(t))/exact),"times more accurate than f1"
````

**Exercise 3 (10 pts)**. Importance of using efficient algorihtms.

- Implement bubble sort
- Calculate its big $\mathcal{O}$ algorithmic complexity
- Time the performance of bubble sort on random uniform deviate vectors of sizes `range(100, 2000, 100)` using `time.time()` from the standard library
- Use `scipy.optimize.curve_fit` to fit an appropriate function to the collection of (size, execution time) data points. Extrapolate how long it would take to sort a random vector of size 1,000,000. Now time how long it takes for the system sort to sort a random vector of size 1,000,000.
- Plot the fits together with the data points uisng `matplotlib.pyplot` functions.

````python
# Your code here

def bubble_sort(a):
    #Conver to numpy array
    a = np.array(a)
    #Creat a shifted array to check whether a is sorted or not
    b = np.append(a[1:len(a)],a[-1])
    #bubble sort
    while np.any((b-a)<0):
        for i in range(len(a)-1):
            if a[i] > a[i+1]:
                #swap
                a[i], a[i+1] = a[i+1], a[i]
        #Creat a shifted array to check whether a is sorted or not
        b = np.append(a[1:len(a)],a[-1])
    return(a.tolist())
#test
ls = [3,6,1,5,3,9,4,0,1,6,3,8]
print bubble_sort(ls)
````

````python
%%latex
The worst case for bubble sort is that the list is totally reversed.
\begin{eqnarray*}
\ \text{Calculation times}  &=& (n-1) + (n-2) + \cdots + 1
\\ &=& \frac{n(n-1)}{2}
\\ &\propto& (n^2 - n)
\end{eqnarray*}

Thus the algorithmic complexity is $\mathcal{O} (n^2)$
````

````python
import time
T1=[]
for i in range(100,2000,100):
    #create random sample for specific sample size
    rls = np.random.uniform(size=i)
    #calculate the time bubble sort used
    t1 = time.time()
    bubble_sort(rls)
    t2 = time.time()
    #store the time 
    T1.append(t2-t1)

print T1
````

**Exercise 4 (20 pts)**. One of the goals of the course is that you will be able to implement novel algorihtms from the literature. 

- Implement the mean-shift algorithm in 1D as described [here](http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/TUZEL1/MeanShift.pdf). 
    - Use the following function signature
    ```python
    def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    ```
    - xs is the data set, x is the starting location, and kernel is a kernel function
    - tol is the difference in $||x||$ across iterations
- Use the following kernels with bandwidth $h$ (a default value of 1.0 will work fine)
    - Flat - return 1 if $||x|| < h$ and 0 otherwise
    - Gaussian 
    $$\frac{1}{\sqrt{2 \pi h}}e^{\frac{-||x||^2}{h^2}}$$
    - Note that $||x||$ is the norm of the data point being evaluated minus the current value of $x$
- Use both kernels to find all 3 modes of the data set in `x1d.npy`
- Modify the algorithm and/or kernels so that it now works in an arbitrary number of dimensions.
- Use both kernels to find all 3 modes of the data set in `x2d.npy`
- Plot the path of successive intermediate solutions of the mean-shift algorithm starting from `x0 = (-4, 10)` until it converges onto a mode in the 2D data for each kernel. Superimpose the path on top of a contour plot of the data density.

````python
# Your code here

def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    #1-D
    if np.ndim(x) == 0:
        #get the kernel function value of x - xi
        grad = map(kernel,xs-x)
        #calculate the mean shift vector
        m = sum(xs*grad)/sum(grad) - x 
        #update x
        y = x + m
        #use a while loop to get the mode
        while np.linalg.norm(y-x) > tol:
            x = y
            #get the kernel function value of x - xi
            grad = map(kernel,xs-x)
            #calculate the mean shift vector
            m = sum(xs*grad)/sum(grad) - x
            #update x
            y = x + m
    #2 or higher dimensions
    else:
        #get the kernel function value of x - xi
        grad = np.array(map(kernel,xs-x))
        #calculate the mean shift vector
        m = sum(grad[:,np.newaxis] * xs)/sum(grad) - x
        #update x
        y = x + m
        #use a while loop to get the mode
        while np.linalg.norm(y-x) > tol:
            x = y
            #get the kernel function value of x - xi
            grad = np.array(map(kernel,xs-x))
            #calculate the mean shift vector
            m = sum(grad[:,np.newaxis] * xs)/sum(grad) - x
            #update x
            y = x + m
    return(y)

    
#bandwidth    
h = 1.0
X1 = np.load('x1d.npy')
X2 = np.load('x2d.npy')
def kern1(x,h = 1.0):
    return(1 if np.linalg.norm(x)**2 < h else 0)

def kern2(x,h = 1.0):
    return(1.0/np.sqrt(2*np.pi*h)*np.exp(- np.linalg.norm(x)**2/(h**2) ))
#the mode of 1D data from flat kernel
m11 = [mean_shift(X1,s,kern1) for s in X1[[10,1600,2600]]]
#the mode of 1D data from Gaussian kernel
m12 = [mean_shift(X1,s,kern2) for s in X1[[10,1600,2600]]]
#the mode of 2D data from flat kernel
m21 = [mean_shift(X2,s,kern1) for s in X2[[10,1200,2000]]]
#the mode of 2D data from Gaussian kernel
m22 = [mean_shift(X2,s,kern2) for s in X2[[10,1200,2000]]]
````

````python
#Gaussian kernel path
tol = 1e-6
#starting point
x0 = np.array([[-4,10]])
#set up the path
path2 = x0
#get the kernel function value of x - xi
grad = np.array(map(kern2,X2-x0))
#calculate the mean shift vector
m = sum(grad[:,np.newaxis] * X2)/sum(grad) - x0
#update x
y = x0 + m
#update path
path2 = np.vstack((path2,y))
#use a while loop to get it
while np.linalg.norm(y-x0) > tol:
    x0 = y
    grad = np.array(map(kern2,X2-x0))
    m = sum(grad[:,np.newaxis] * X2)/sum(grad) - x0
    y = x0 + m  
    path2 = np.vstack((path2,y))

print path2
````
